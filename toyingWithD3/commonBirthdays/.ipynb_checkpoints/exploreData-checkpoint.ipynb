# We need to calculate the average
- what consist the average?
- the average equals the average daily live-births for 23 years
- that means, we need to some the daily data for 23 years, and divide

```python
# this variable holds the mean of yearly daily datas, which means, how many live-births occured on average daily
# during the 23 year period
meanForYears = dataset.mean(axis=1, numeric_only=True, skipna=True)
```

```python
# after that, we calculate another mean, the average of daily births of the before-calculated averages
mean = meanForYears.mean()
```

```python
# finalRatios, that we going to visualize
finalRatios = meanForYears / mean
```

```python
# final data to visualize, then we make a dataframe out of it, to export as json
finalData = {
    'date': dataset['Date'],
    'mean': mean,
    'dailyAverages': meanForYears,
    'ratio': finalRatios
}
```

```python
finalDataset = pd.DataFrame(finalData)
```
